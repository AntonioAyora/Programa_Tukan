````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import glob
````

````python
files = glob.glob('datos/Ventas/*.txt')
df =  pd.concat((pd.read_csv(file) for file in files))
df=df.dropna(axis='rows',how='any')
df=df[df['Quantity Ordered']!='Quantity Ordered'] 
df
````

<div class="alert alert-success">
    
1. Productos que se adquirieron en Los Angeles, CA
    
</div> 

````python
def direcc(x):
    a = x.split(',')[0]
    return a
def city(x):
    b = x.split(',')[1]
    return b
def cp(x):
    c = x.split(',')[2]
    return c
df["Address"]=df["Purchase Address"].apply(direcc)
df["City"]=df["Purchase Address"].apply(city)
df["CP"]=df["Purchase Address"].apply(cp)
df.head()
````

````python
pro = (df[df['City'] == ' Los Angeles'])['Product'].count()
print('Cantidad de productos que se adquirieron en Los Angeles, CA =', pro)
````

<div class="alert alert-success">
    
2. Iphone vendidos

    
</div> 

````python
df[['Quantity Ordered', 'Price Each']] = df[['Quantity Ordered', 'Price Each']].apply(pd.to_numeric)
````

````python
pho = (df[df['Product'] == 'iPhone'])['Quantity Ordered'].sum()
print('Cantidad de iPhone que se vendieron=', pho)
````

<div class="alert alert-success">
    
3. Ganancias por venta de baterías 
    
</div> 

````python
(df[df['Product'] == 'AA Batteries (4-pack)'])
````

````python
a = df.groupby('Product')['Quantity Ordered'].sum()
b=((a['AA Batteries (4-pack)']*3.84)+(a['AAA Batteries (4-pack)']*2.99)).round(2)
print('Ganancia por venta de baterías = ',b)
````

<div class="alert alert-success">
    
4. Código postal con más ordenes realizadas?

    
</div> 

````python
z = df.groupby('CP')['Quantity Ordered'].sum().reset_index().sort_values('Quantity Ordered', ascending=False)
a=z.iloc[0][0].split()[1]
print('código postal que más ordenes realizó = ', a)
````

<div class="alert alert-success">
    
5.a  Producto con más ganancias

    
</div> 

````python
df['Ganancias'] = df['Quantity Ordered']*df['Price Each']
z = df.groupby('Product')['Ganancias'].sum().reset_index().sort_values('Ganancias', ascending=False)
a=z.iloc[0][0]
print('Producto con más ganancias = ', a)
````

<div class="alert alert-success">
    
5.b Producto con más unidades vendidas

    
</div> 

````python
g = df.groupby('Product')['Quantity Ordered'].sum().reset_index().sort_values('Quantity Ordered', ascending=False)
a=g.iloc[0][0]
print('Producto con más unidades vendidas = ', a)
````

<div class="alert alert-success">
    
8. Ventas Totales Por Hora   
</div> 

````python
df['Fecha'] = pd.to_datetime(df['Order Date'], infer_datetime_format=True)
df = df.set_index('Fecha')
df
````

````python
a = df['Quantity Ordered'].groupby(df.index.hour).sum()
a.plot(style = ['r--'],figsize =(10,5))
plt.xlabel('Numero de ventas')
plt.ylabel('Hora del Dia')
plt.title("Ventas total en Cada Hora")
plt.show()
````

<div class="alert alert-success">
    
 9. Día con mas ganacias y con menos ganancias
    
</div> 

````python
a = df['Quantity Ordered']
b= a[a.index.month.isin([1])]
b1= a[a.index.month.isin([2])]
c=(b.groupby(b.index.day).sum().reset_index().sort_values('Quantity Ordered', ascending=False)).iloc[0][0]
c1=(b1.groupby(b1.index.day).sum().reset_index().sort_values('Quantity Ordered', ascending=False)).iloc[0][0]
print('Día de más ganancias en enero = ',c)
print('Día de más ganancias en febrero = ',c1)
````

````python
a=df.groupby(df.index.month).sum().reset_index().sort_values("Ganancias", ascending=False)
l=df.groupby(df.index.month).sum().reset_index().sort_values("Ganancias", ascending=True)
b=a.iloc[0][0]
m=l.iloc[0][0]
print(' Mes con más ganancias =', b)
print(' Mes con menos ganancias =', m)
````

<div class="alert alert-success">
    
10. Producto más comprado en San Francisco, CA
    
</div> 

````python
def state(x):
    a = x.split()[0]
    return a
df["State"]=df["CP"].apply(state)
df
````

````python
a=(df[df['City'] == ' San Francisco'])
b=(a[a['State'] == 'CA'])
g = b.groupby('Product')['Quantity Ordered'].sum().reset_index().sort_values('Quantity Ordered', ascending=False)
c=g.iloc[0][0]
print('Producto más comprado en San Francisco, CA = ', c)
````
